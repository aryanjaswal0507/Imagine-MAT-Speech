```python
import numpy as np
from scipy.io import loadmat, savemat

# Load the original .mat file
data = loadmat('/Users/aryanjaswal/Projects/IMAGINE MAT/DATASET/Vowels/sub_4b_ch80_v_eog_removed_256Hz.mat')

# Access the EEG data
eeg_data = data['eeg_data_wrt_task_rep_no_eog_256Hz_last_beep']
print("Original data shape:", eeg_data.shape)

# Initialize a new array with the same shape
trimmed_data = np.empty(eeg_data.shape, dtype=object)

# Loop over all cells and take only the first 64 channels
for i in range(eeg_data.shape[0]):
    for j in range(eeg_data.shape[1]):
        cell = eeg_data[i, j]
        if isinstance(cell, np.ndarray) and cell.ndim >= 1:
            trimmed_data[i, j] = cell[:64]  # Keep first 64 channels
        else:
            trimmed_data[i, j] = cell  # Keep as-is if not an array

# Save to a new .mat file
savemat('/Users/aryanjaswal/Projects/IMAGINE MAT/DATASET/Vowels/sub_4b_ch64_v_eog_removed_256Hz.mat',
        {'eeg_data_wrt_task_rep_no_eog_256Hz_last_beep': trimmed_data})

print("✅ New .mat file with first 64 channels saved.")
```

```python
import numpy as np
from scipy.io import loadmat, savemat

# List of .mat file paths (add your full paths here)
mat_files = [
    '/Users/aryanjaswal/Projects/IMAGINE MAT/DATASET/Long_words/sub_2b_ch64_l_eog_removed_256Hz.mat',
    '/Users/aryanjaswal/Projects/IMAGINE MAT/DATASET/Long_words/sub_3b_ch64_l_eog_removed_256Hz.mat',
    '/Users/aryanjaswal/Projects/IMAGINE MAT/DATASET/Long_words/sub_6_ch64_l_eog_removed_256Hz.mat',
    '/Users/aryanjaswal/Projects/IMAGINE MAT/DATASET/Long_words/sub_7_ch64_l_eog_removed_256Hz.mat',
    '/Users/aryanjaswal/Projects/IMAGINE MAT/DATASET/Long_words/sub_9c_ch64_l_eog_removed_256Hz.mat',
    '/Users/aryanjaswal/Projects/IMAGINE MAT/DATASET/Long_words/sub_11b_ch64_l_eog_removed_256Hz.mat'
]

# Variable key inside each .mat file
mat_key = 'eeg_data_wrt_task_rep_no_eog_256Hz_last_beep'

# List to hold second rows
second_rows = []

# Extract second row (index 1) from each file
for file_path in mat_files:
    data = loadmat(file_path)
    
    if mat_key in data:
        eeg_data = data[mat_key]
        second_row = eeg_data[1:2, :]  # Keep 2D shape (1, 100)
        second_rows.append(second_row)
    else:
        print(f"⚠️ Key '{mat_key}' not found in {file_path}")

# Combine all second rows
combined_data = np.vstack(second_rows)  # Shape: (n_files, 100)

# Save the combined data to a new .mat file
savemat('independent.mat', {mat_key: combined_data})

print("✅ Combined .mat file (2nd rows) saved with shape:", combined_data.shape)
```
